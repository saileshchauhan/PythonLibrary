```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.express as px
```

```python
fig_ScatterPlot = px.scatter(
                            x=np.random.rand(100),
                            y=np.random.randn(100) 
                            )
fig_LinePlot=px.line(x=np.random.randn(100),
                 y=np.random.randn(100))

fig_Combined=go.Figure(data=fig_LinePlot.data+fig_ScatterPlot.data)
```
