```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X=np.random.randint(10,size=15)
Y=np.random.randint(10,size=15)
```

```python
plt.scatter(X,Y,s=150,color='darkblue')
plt.show()
```
