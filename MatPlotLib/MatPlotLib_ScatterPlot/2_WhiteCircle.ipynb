```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X=np.random.randint(10,size=15)
Y=np.random.randint(10,size=15)
sizes = 10*np.random.randint(100, size=(15))
plt.scatter(X, Y, s=sizes,edgecolor='green',facecolor='none')
plt.show()
```
