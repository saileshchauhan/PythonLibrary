```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
X_Axis=random.randint(50,size=5)
```

```python
Y1_Axis=np.log10(X_Axis)
Y2_Axis=np.log2(X_Axis)
Y3_Axis=np.log(X_Axis)
```

```python
axis=plt.subplots(2,2)
plt.figure(figsize=(8,6),dpi=100)
axis[0,0].plot(X_Axis,Y1_Axis)
axis[0, 0].set_title("Log10 Function")
```
