```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xValues=np.arange(0,40,5)
yValues=[value*(3) for value in xValues]
```

```python
plt.plot(xValues,yValues)

plt.xticks(np.arange(0,50, 5))
plt.yticks(np.arange(0,120,20))

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Draw a line graph')

plt.show()
```
