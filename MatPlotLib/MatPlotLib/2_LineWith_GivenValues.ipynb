```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xValues=[1,2,3,4,5]
yValues=[4,6,7,2,4]
```

```python
plt.plot(xValues,yValues)

scale_factor_x=1.5
scale_factor_y=1.5

xmin,xmax=plt.xlim()
ymin,ymax=plt.ylim()

plt.xlim(xmin,xmax*scale_factor_x)
plt.ylim(xmin,xmax*scale_factor_y)

plt.xlabel('X-aixs')
plt.ylabel('Y-axis')
plt.title('line graph X and Y')

plt.show()
```
