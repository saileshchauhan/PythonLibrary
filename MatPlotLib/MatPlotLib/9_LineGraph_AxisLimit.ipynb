```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xValues=np.arange(0,150,5)
yValues=[value*(3) for value in xValues]
```

```python
plt.plot(xValues,yValues,label='Line 1',color='skyblue',linewidth=4,marker='o',mfc='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two lines with current axis limit values')
print(plt.axis((0,100,0,200)))
plt.legend()
plt.show()
```
