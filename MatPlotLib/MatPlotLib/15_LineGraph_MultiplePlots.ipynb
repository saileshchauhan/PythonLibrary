```python
import matplotlib.pyplot as plt
from numpy import random
```

```python
fig=plt.figure(figsize=(8,6),dpi=100)

X_Plot1=random.randint(50,size=5)
Y_plot1=random.randint(30,size=5)
plt.subplot(2,1,1)
plt.plot(X_Plot1,Y_plot1)

X_Plot2=random.randint(40,size=6)
Y_plot2=random.randint(35,size=6)
plt.subplot(2,1,2)
plt.plot(X_Plot2,Y_plot2)
plt.show()
```
