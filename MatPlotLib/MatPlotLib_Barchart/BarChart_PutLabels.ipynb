```python
'''
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
'''
```

```python
import matplotlib.pyplot as plt
```

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i,x in enumerate(x)]

plt.bar(x_pos, popularity, color=(0.4, 0.6, 0.8, 1.0),edgecolor='blue')
fig, axis = plt.subplots()
rectangles = axis.bar(x_pos, popularity, color='b')

plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago")
plt.xticks(x_pos, x)

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')

plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
def autolabel(rectangles):
    '''
    Description:
        Provides label above each bar.
    Parameters:
        Takes list of bar(rectangle).
    Returns:
        None.
    '''
    for rect in rectangles:
        height = rect.get_height()
        axis.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%f' % float(height),
        ha='center', va='top')
autolabel(rectangles)
plt.show()
```
